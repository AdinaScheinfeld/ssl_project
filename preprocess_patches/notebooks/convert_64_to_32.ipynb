# Notebook to convert 64 bit image to 32 bits
ITK Snap cannot view 64 bit images, so this notebook converts 64 bit images from CellSeg3D to 32 bits  
Data: https://zenodo.org/records/11095111  
6 volumes were annotated: 5 from somatosensory cortex (clean), 1 from visual cortex (with some imaging artifacts)

```python
# imports
import tifffile as tiff
import numpy as np
```

```python
# get list of 64 bit labels
labels = [
    'c1labels_new_label',
    'c2labels_new_label',
    'c3labels_new_label',
    'c4labels_new_label',
    'c5labels_new_label',
    'visual_gt_new_label'
]
```

```python
# loop over labels and conver each one to 32 bit
for label in labels:
    label_path = f'/midtier/paetzollab/scratch/ads4015/cellseg3d_data/labels/{label}.tif'
    label_64 = tiff.imread(label_path)
    label_32 = label_64.astype(np.float32)
    tiff.imwrite(f'/midtier/paetzollab/scratch/ads4015/cellseg3d_data/labels/{label}_32bit.tif', label_32)
    print(f'Converted {label} to 32 bit and saved as {label}_32bit.tif')
```
